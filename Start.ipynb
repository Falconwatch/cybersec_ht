# Проанализируем примеры
Посмотрим какие поля чаще всего различаются в данных от разных поставщиклов

```python
unique_cols = list(set([c[3:] for c in examples.columns]))
unique_cols
```

```python
same_ratios = list()
for c in unique_cols:
  t1 = examples["c1."+c]
  t2 = examples["c2."+c]
  examples[c+"_similar"] = (t1==t2)
  same_ratio = (t1==t2).mean()
  same_ratios.append((c, same_ratio))
  
```

# Вычислим насколько отличаются два поля

```python
from difflib import SequenceMatcher
```

```python
def similarity_ratio_func(df, field):
  name1 = "c1."+field
  name2 = "c2."+field
  name3 = field+"_similarity_ratio"
  df[name3] = df[[name1, name2]].apply(lambda x: 
                  SequenceMatcher(None, str(x[name1]), str(x[name2])).ratio(), axis=1)
```

```python
for c in unique_cols:
  similarity_ratio_func(examples, c)
```

```python
examples[[c for c in examples.columns if "similarity_ratio" in c]].mean()
```

```python
examples[[c for c in examples.columns if "similarity_ratio" in c]].std()
```
